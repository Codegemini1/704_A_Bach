```python
def even_odd(x):
    if x % 2 == 0:
        return 'Even'
    else:
        return 'Odd'
```

```python
df['Even_Odd'] = df['Age'].apply(even_odd)
```

```python
df.head()
```

```python
df['Square'] = df['Pclass'].apply(lambda x : x * x)
```

```python
df.head()
```

```python
def gender(x):
    if x == 'male':
        return 1
    else:
        return 0
```

```python
df['Gender'] = df['Sex'].apply(gender)
```
