```python
from nltk.corpus import stopwords

words = ["i", "love", "learning", "python"]
filtered = [w for w in words if w not in stopwords.words("english")]
print(filtered)
```

```python
text = 'Today I am very happy' #Remove stowords 
```

```python
text = text.split()
```

```python
text
```

```python
for i in text:
    print(i)
```

```python
for word in text:
    if word in stopwords.words('english'):
        print(word, '-:Found in stopword')
    else:
        print(word,'-:Not in Stopward')
```

```python
text
```

```python
for word in text:
    if word in stopwords.words('english'):
        text.remove(word)
print(text)
```
