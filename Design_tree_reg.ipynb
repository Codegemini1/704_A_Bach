```python
from sklearn.tree import DecisionTreeRegressor
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {
    'Area': [800,1000,1200,2000,2200],
    'Price': [30,35,40,70,75]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
X = df[['Area']]
y = df[['Price']]
```

```python
model = DecisionTreeRegressor(max_depth = 2, random_state= 0 )
```

```python
model.fit(X,y)
```

```python
test = np.arange(min(X.Area), max(X.Area), 10).reshape(-1,1)
```

```python
test
```

```python
y_pred = model.predict(test)
```

```python
plt.scatter(X,y, color = 'red', label ='Actual Prices')
plt.plot(test, y_pred, color = 'blue', label = 'Predicted Prices (Tree)')
plt.title('Design Tree Reg - House Price Pred')
plt.xlabel('Area(sq.ft)')
plt.ylabel('Prices $')
plt.legend()
plt.show()
```
