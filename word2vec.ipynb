```python
from gensim.models import Word2Vec
```

```python
sentences = [
    ['i', 'love', 'nlp'],
    ['i', 'love', 'python'],
    ['nlp', 'is', 'powerful']
]
```

```python
model = Word2Vec(sentences , vector_size=50, window=2, min_count=1)
```

```python
model.wv['nlp']
```

```python
model.wv['python']
```
